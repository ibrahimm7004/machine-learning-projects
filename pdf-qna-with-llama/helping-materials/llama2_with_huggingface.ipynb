```python
import torch
torch.cuda.is_available()
```

```python
from transformers import AutoTokenizer,AutoModelForCausalLM
import transformers

access_token = ""
model = "meta-llama/Llama-2-7b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model, token=access_token)

model = AutoModelForCausalLM.from_pretrained(
    model,
    token=access_token
)

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
sequences = pipeline(
    'Hi! Tell me about yourself!',
    do_sample=True,
)
print(sequences[0].get("generated_text"))
```

```python
# Step 1 : Import and create bit and bytes config

from torch import cuda, bfloat16
from transformers import BitsAndBytesConfig

bnb_config = transformers.BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=bfloat16
)
```

```python
#Step 2 : Loading model with quantization config
```

```python
model = AutoModelForCausalLM.from_pretrained(
    "./models/llama2",
    device_map={"": 0},
    quantization_config=bnb_config,
)

model.config.use_cache = False
model.config.pretraining_tp = 1
```

```python
#Next, let’s create a pipeline with the model loaded using the quantization configuration and test the speed once again.

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
)

sequences = pipeline(
    'Hi! Tell me about yourself!',
    do_sample=True
)
```
